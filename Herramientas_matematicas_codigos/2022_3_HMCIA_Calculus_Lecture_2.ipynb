##Increasing and Decreasing functions

Let $f$ be a function defined on an interval, and $x_1{ y } x_2$ points on the interval.
*   $f$ is increasing on the interval if $f(x_1)< f(x_2)$ whenever $x_1<x_2$
*   $f$ is decreasing on the interval if $f(x_1)>f(x_2)$ whenever  $x_1<x_2$
*   $f$ is constant on the interval if $f(x_1)=f(x_2)$ for every point $x_1$ and $x_2$


image.png

Let $f$ be a function that is continuous on a closed interval $[a,b]$ and differentiable on the open interval $(a,b)$.

*   If $f'(x)>0$ for every value of $x$ in $(a,b)$, then $f$ is increasing in $[a,b]$.
*   If $f'(x)<0$ for every value of $x$ in $(a,b)$, then $f$ is decreasing in $[a,b]$.
*   If $f'(x)=0$ for every value of $x$ in $(a,b)$, then $f$ is constant in $[a,b]$.

##Optimizacion

The solution of optimization problems is one of the main applications of differentiation. 

An optimization problem refers to finding the maximum or minimum value of a function.

Global minimum: $x^* \in S, \text{ if } f (x^*) \leq f (x)
\text{ for all }  x \in S$.

Local minimum: $x^* \in S, \text{ if } f (x^*) \leq f (x)
\text{ for all } x$ within an infinitesimally small distance $\epsilon$ of $x^*$. That is, there exists $\epsilon > 0$ such that for all $x$ satisfies $|x-x^*| < \epsilon, f(x^*)\leq f(x)$.


Optimization theory provides the tools to find a "good" value of $x$ that corresponds to a "good" value of $f(x)$.


image.png

##Necessary Conditions for a Local Optimum

Assuming that the first and second derivatives of $f(x)$ exist, the necessary conditions for $x^*$ to be a local minimum/maximum of the function $f(x)$ on an interval $(a,b)$ are:

**Local Minimum**

$\frac{df}{dx}\Bigr|_{x=x^*}=0$

$\frac{d^2f}{dx^2}\Bigr|_{x=x^*}\geq 0$


**Local Maximum**

$\frac{df}{dx}\Bigr|_{x=x^*}=0$


$\frac{d^2f}{dx^2}\Bigr|_{x=x^*}\leq 0$


These conditions are necessary, but not sufficient. That they are fulfilled does not guarantee that $x^*$ is a local minimum or maximum.





##Sufficient Conditions for a Local Optimum

Let $x^*$ be a point at which the first derivative is zero and the order of the first nonzero higher derivative is $n$.

*   If $n$ is odd, then $x^*$ is an inflection point.
*   If $n$ is even, then $x^*$ is a local optimum. Moreover:
   *   If the value of the derivative of $f''(x)$ at $x^*$ is positive, then the point $x^*$ is a local minimum.
   *   If the value of the derivative of $f''(x)$ at $x^*$ is negative, then the point $x^*$ is a local maximum.

###Finding a 3-Point Interval

If your goal is to optimize (maximize or minimize) the value of an equation, one of the very first steps is to find an interval where that value might occur. This is obvious if you know the graph… but what if you don't?

image.png

###The first step

The first thing you do is find a point. Any point. Since you have no clues about the graph or its shape, you’re flying blind. Pick your favorite number and plug it in.
Notice that we still have no information about the shape of the graph.


image.png

###The next step

Next, you find a second point pretty close to the first. You can do this by guessing a second number, but you can save steps later by just picking an interval value, which I’ll call h. your second point will have an x-coordinate h units larger than the first.

So if you chose x = 2 and h = 0.5, you’ll have points at (2, f(2)) and (2.5, f(2.5)).

image.png

###The Initial Test

Once you have the second point, you compare it to the first. 

If you’re seeking a minimum, you would like your second point to be lower (smaller y) than the first.

If you’re seeking a maximum, you want your second point to be higher (larger y).


image.png

If the second point fails the test, change your h to –h and try again with (x, f(x)) and (x-h, f(x-h)).

If the graph is decreasing to the right, it’s usually increasing to the left, and vice-versa. 

image.png

###After the initial test

Once you know your direction, take another step in that direction using the same value of h. 

If you were minimizing a decreasing function it might look like this:

Now, you have 3 points.

###The Goal

If you’re looking for a minimum, your ultimate goal is to find an interval of three points such that the point in the middle is lower than both endpoints:

When you get this type of picture, you know there is a minimum somewhere in the interval between xk and xk+2.

###The problem with h

If h is too big, your margin of error is large and you risk skipping over the maximum or minimum completely.

If h is too small, the calculations become too numerous.



##Problems with the procedure

*  This procedure will only find a local minimum or maximum, not necessarily the absolute minimum or maximum.

*  Also, if the function is unbounded, you might not find an interval at all.

*  There is no easy way to prevent the first error (local vs. absolute), but some advanced methods have been developed to do so.

*  For the second error (unboundedness), you can either introduce a boundary (like x must be between - and - ), or you can stop the process if you appear to be heading off towards infinity, and start again with a different x.

###Local vs. Global Extremes

If a point is a maximum or minimum relative to the other points in its “neighborhood”, then it is considered a local maximum or local minimum.
Local vs. Global Extremes

###Unimodal and Multimodal 

A unimodal function has only one minimum and the rest of the graph goes up from there; or one maximum and the rest of the graph goes down.

With unimodal functions, any extreme you find is guaranteed to be the global extreme.

A bimodal function has two local minima or maxima.

Beyond that, trimodal, quadrimodal and then multimodal.

###Refining the Program

The solution to the problem with h is to start with a small h and make it larger as the loops count up. Our modified loop was this:
1.   Start with (x, f(x)).
2.   Find (x+h, f(x+h)) and check if it’s going the right direction. If not, find (x-h, f(x-h)).
3.   Take one more step h or –h.
4.   Check if you have reached the goal (middle point higher/lower than both endpoints). 
5.   If yes, stop.
6.   If no, increase h and return to step 3.

#Gradient and Hessian Matrix


##Gradient of a Function.

Given a function $f(x)$, its gradient is:

\begin{align}
\nabla (f(x))=
\begin{bmatrix}
  \frac{\partial f}{\partial x_1}\\
  \vdots\\
  \frac{\partial f}{\partial x_n}\\
\end{bmatrix}
\end{align}


where $x$ is an n-dimensional vector.

**Example:**

Given the function $f(x)=2x_1^4+4x_2^3-3x_1x_2^2$ calculate the gradient.

**Solution:**
\begin{align}
  \nabla (f(x))=
  \begin{bmatrix}
  8x_1^3+3x_2^2\\
  12x_2^2+6x_1x_2
\end{bmatrix}
\end{align}


The gradient is a vector of $n \times 1$.


image.png

```python
x1, x2 = syp.symbols('x1 x2')
f = 2*x1**4+4*x2**3+3*x1*x2**2
gradient=[[syp.diff(f,x1)],[syp.diff(f,x2)]]
print(gradient)
f_J=syp.Matrix([f]).jacobian(syp.Matrix(list(f.free_symbols)))
f_J
```

###Evaluating the Gradient

*  In 1-variable calculus, the derivative gives you an equation for the slope at any x-value along f(x). You can then plug in an x-value to find the actual slope at that point.

*  Similarly, the gradient gives you an equation for the slope of the tangent plane at any point (x, y) or (x, y, z) or whatever. You can then plug in the actual values at any point to find the slope of the tangent plane at that point. 

*  The slope of the tangent plane will be written as a vector, composed of the slopes along each dimension.

*  **The direction of the gradient vector will always point in the direction of steepest increase for the function and, its magnitude will give us the slope of the plane in that direction**.

image.png

##Hessian Matrix

Just as the first derivative in 2 or more variables has a special name, gradient, the second also has a special name, Hessian, after the developer, Ludwig Otto Hesse.

The Hessian of a function $f(x)$ is given by:

\begin{align}
\nabla^2(f(x))=H(f)=
\begin{bmatrix}
  \frac {\partial ^2f} {\partial x_1^2} & 
  \frac {\partial ^2f} {\partial x_1 \partial x_2} &
  \cdots &
  \frac {\partial ^2f} {\partial x_1 \partial x_n} \\
  \frac {\partial ^2f} {\partial x_2 \partial x_1} &
  \frac {\partial ^2f} {\partial x_2^2} &
  \cdots &
  \frac {\partial ^2f} {\partial x_2 \partial x_n}\\
  \vdots & \vdots & \ddots & \vdots \\
  \frac {\partial ^2f} {\partial x_n \partial x_1}&
  \frac {\partial ^2f} {\partial x_n \partial x_2}&
  \cdots &
  \frac {\partial ^2f} {\partial x_n^2}
\end{bmatrix}
\end{align}

The Hessian matrix is symmetric.


**Example:**  

Given the function $f(x)=2x_1^4+4x_2^3-3x_1x_2^2$ calculate the Hessian.

Solution:
\begin{align}
\nabla^2(f(x))=H(f)=
\begin{bmatrix}
24x_1^2 & -6x_2\\\\
-6x_2 & 24x_2-6x_1
\end{bmatrix}	
\end{align}


```python
x1, x2 = syp.symbols('x1 x2')
f = 2*x1**4+4*x2**3-3*x1*x2**2
hessian=[[syp.diff(f,x1,2), syp.diff(f,x1,x2)],[syp.diff(f,x2,x1), syp.diff(f,x2,2)]]
print(hessian)
f_H=syp.hessian(syp.Matrix([f]), syp.Matrix(list(f.free_symbols)))
f_H
```

Because the Hessian of an equation is a square matrix, its **eigenvalues** can be found

Because Hessians are also **symmetric** (the original and the transpose are the same), they have a special property that their eigenvalues will always be **real numbers**.

So the only thing of concern is whether the eigenvalues are **positive** or **negative**.

If the **Hessian at a given point has all positive eigenvalues**, it is said to be a **positive-definite matrix**. This is the multivariable equivalent of **“concave up”**.

image.png

If **all of the eigenvalues are negative**, it is said to be a **negative-definite matrix**. This is like **“concave down”**.


image.png

If **either eigenvalue is 0**, then you will **need more information** (possibly a graph or table) to see what is going on.

And, if the **eigenvalues are mixed (one positive, one negative)**, you have a saddle point. Here, the graph is concave up in one direction and concave down in the other.

image.png

##Scipy.Optimize Library

SciPy optimize provides functions for minimizing (or maximizing) objective functions, possibly subject to constraints. It includes solvers for nonlinear problems (with support for both local and global optimization algorithms), linear programing, constrained and nonlinear least-squares, root finding, and curve fitting.

**Scalar functions optimization**

`minimize_scalar(fun[, bracket, bounds, …])`

Minimization of scalar function of one variable.


**Local (multivariate) optimization**

`minimize(fun, x0[, args, method, jac, hess, …])`

Minimization of scalar function of one or more variables.

Constraints are passed to minimize function as a single object or as a list of objects from the following classes:

`NonlinearConstraint(fun, lb, ub[, jac, …])`
`LinearConstraint(A, lb, ub[, keep_feasible])`


**Global optimization**

`basinhopping(func, x0[, niter, T, stepsize, …])`

Find the global minimum of a function using the basin-hopping 

`brute(func, ranges[, args, Ns, full_output, …])`

Minimize a function over a given range by brute force.

`differential_evolution(func, bounds[, args, …])`

Finds the global minimum of a multivariate function.

`shgo(func, bounds[, args, constraints, n, …])`

Finds the global minimum of a function using SHG optimization.

`dual_annealing(func, bounds[, args, …])`

Find the global minimum of a function using Dual Annealing.


**Least-squares and curve fitting**

Nonlinear least-squares:    
`least_squares(fun, x0[, jac, bounds, …])`

Solve a nonlinear least-squares problem with bounds on the variables.

Linear least-squares: 
`nnls(A, b[, maxiter]) Solve argmin_x || Ax - b ||_2 for x>=0.`

`lsq_linear(A, b[, bounds, method, tol, …])`

Solve a linear least-squares problem with bounds on the variables.


**Curve fitting**

`curve_fit(f, xdata, ydata[, p0, sigma, …])`

Use non-linear least squares to fit a function, f, to data.


**Linear programming**

`linprog(c[, A_ub, b_ub, A_eq, b_eq, bounds, …])`

Linear programming: minimize a linear objective function subject to linear equality and inequality constraints.

```python
minb=-2.5
maxb=2.5
result = optimize.minimize_scalar(scalar1, bounds = (minb,maxb),method='Bounded')
print("When bounded between ",minb," and ",maxb, " minimum occurs at: ",result['x'])
x=np.arange(-10,10,0.1)
y=scalar1(x)
plt.plot(x,y)
print(scalar1((result['x'])))
plt.plot(result['x'],scalar1((result['x'])),marker="s")
```

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

np.random.seed(0)

# Our test function
def f(t, omega, phi):
    return np.cos(omega * t + phi)

# Our x and y data
x = np.linspace(0, 3, 50)
y = f(x, 1.5, 1) + .1*np.random.normal(size=50)

# Fit the model: the parameters omega and phi can be found in the
# `params` vector
params, params_cov = optimize.curve_fit(f, x, y)

# plot the data and the fitted curve
t = np.linspace(0, 3, 1000)

plt.figure(1)
plt.clf()
plt.plot(x, y, 'bx')
plt.plot(t, f(t, *params), 'r-')
plt.show()
```

It may be desirable to maximize the final resultant process output by choosing the optimum operating points in the individual sub-processes (within certain process limits).

Because the optimization problem here is about maximization of the objective  function, we need to change the sign and return the negative of the sum of the Gaussian functions as the result of the objective function.

```python
def f1(x):
     return np.exp(-(x+1)**2/(2.1**2))
def f2(x):
     return np.exp(-(x-0.3)**2/(0.8**2))
def f3(x):
     return np.exp(-(x-2.1)**2/(1.7**2))
  

x=np.arange(-5,5,0.1)
plt.plot(x,f1(x),'r')
plt.plot(x,f2(x),'g')
plt.plot(x,f3(x),'b')

plt.plot(x,f1(x)+f2(x)+f3(x),'k--')


```

```python
import numpy as np
from scipy import optimize
def gaussian_mixture(x):
    """
    Computes the resultant Gaussian mixture from an input vector and 
    known mean, variance quantities
    """
    return -(f1(x[0])+
             f2(x[1])+
             f3(x[2]))
x0=np.array([0]*3)
result = optimize.minimize(gaussian_mixture,x0=x0,method='SLSQP',
                           options={'maxiter':100})
print(result['x'])
```

```python
import numpy as np
from scipy.optimize import minimize, LinearConstraint

n_buyers = 10
n_shares = 15

prices = np.array([0.77, 0.02, 0.63, 0.75, 0.5, 0.22, 0.2, 0.76, 0.17, 0.09])
money_available = np.array([1, 1, 1, 3, 1, 3, 3, 2, 1, 1])

n_shares_per_buyer = money_available / prices
print(prices, money_available, n_shares_per_buyer, sep="\n")


```

The constraint is that the sum of the total purchased shares can’t exceed the total number of shares available. This is a constraint rather than a bound because it involves more than one of the solution variables.

To represent this mathematically, you could say that x[0] + x[1] + ... + x[n] = n_shares, where n is the total number of buyers. 

```python
def constraint1(x):
    return np.sum(x)-15
constraint = {'type':'eq','fun':constraint1}
```

We could also write: <br>
constraint = LinearConstraint(np.ones(n_buyers), lb=n_shares, ub=n_shares)
<br>
¿How does that work?

The bounds limit the number of shares purchased to be 0 on the lower side and n_shares_per_buyer on the upper side. The format that minimize() expects for the bounds is a sequence of tuples of lower and upper bounds

```python
bounds = [(0, n) for n in n_shares_per_buyer]
print(bounds)
```

Define objective_function() to take two arguments. Then take the dot product of x with prices and return the negative of that value. You have to return the negative because you’re trying to make that number as small as possible, or as close to negative infinity as possible

```python


def objective_function(x, prices):
    #return -x.dot(prices)
    return -np.dot(x, prices)

#x0=10 * np.random.random(n_buyers),
x0=np.zeros(n_buyers)
res = minimize(
    objective_function,
    x0,
    args=(prices,),
    constraints=constraint,
    bounds=bounds,
)

print("The total number of shares is:", sum(res.x))
print("Leftover money for each buyer:", np.round(money_available - res.x * prices, decimals=2))
print("Number of shares sold to each buyer:",np.round(res.x))
print("Your total income is $",np.round(-res.fun, decimals=2))
```

##Learning rate 

The size of these steps is called the learning rate (α) that gives us some additional control over how large of steps we make. With a large learning rate, we can cover more ground each step, but we risk overshooting the lowest point since the slope of the hill is constantly changing. With a very low learning rate, we can confidently move in the direction of the negative gradient since we are recalculating it so frequently. A low learning rate is more precise, but calculating the gradient is time-consuming, so it will take us a very long time to get to the bottom.

There are several variants of the Gradient Descent algorithm that try to address this problems

###Minimizing with Steepest Descent

The Steepest Descent Method for minimization begins with the idea that the gradient is a vector pointing in the direction of fastest increase for the function. The steepest descent method uses the negative value of the gradient to find the ideal direction of decrease. 

The process continues until a condition is met – often one or a combination of: 
* a number of iterations
* a low enough value of the gradient (which approaches 0 as the optimal point is neared)
* a low enough change in the vector 
* a low enough change in the function value between the old and new points

###Conjugate Gradient Method

The conjugate gradient method takes into account information from directions of previous gradients to converge faster twoards the optimum

image.png

```python
import numpy as np
from numpy import linalg as LA


A=np.array([[3,2],[2,3]])
B=np.array([-2,7])
x0=np.array([0,0])

def is_pos_def(x):
    """check if a matrix is symmetric positive definite"""
    return np.all(np.linalg.eigvals(x) > 0)

def steepest_descent(A, b, x):
    """
    Solve Ax = b
    Parameter x: initial values
    """
    if (is_pos_def(A) == False) | (A != A.T).any():
        raise ValueError('Matrix A needs to be symmetric positive definite (SPD)')
    r = b - A @ x
    k = 0
    while LA.norm(r) > 1e-10 :
        p = r
        q = A @ p
        alpha = (p @ r) / (p @ q)
        x = x + alpha * p
        r = r - alpha * q
        k =+ 1

    return x

def conjugate_gradient(A, b):
    if (is_pos_def(A) == False) | (A != A.T).any():
        raise ValueError('Matrix A needs to be symmetric positive definite (SPD)')
    r = b 
    k = 0
    x = np.zeros(A.shape[-1])
    while LA.norm(r) > 1e-10 :
        if k == 0:
            p = r
        else: 
            gamma = - (p @ A @ r)/(p @ A @ p)
            p = r + gamma * p
        alpha = (p @ r) / (p @ A @ p)
        x = x + alpha * p
        r = r - alpha * (A @ p)
        k =+ 1
    return x
```


##Minimization with Nonlinear Constraints: 

###Direction Requirements
The first step in many nonlinear optimization techniques is to start on the boundary of the feasible region – it doesn’t matter which boundary or where on the boundary you start.

There are two considerations when choosing what direction to move next. If we are minimizing, any vector we choose must: 
*  decrease the objective function 
*  stay in the feasible region



###Conditions for an Optimal Point
Another possibility is that the optimal value occurs in the interior of the feasible region rather than on a boundary. 

In this case the gradient of the objective function will equal zero at that point, just as the derivative equals zero at optimal points on 1-variable functions. 

However, a gradient value of 0 is not enough to guarantee an optimal value: you must still check the eigenvalues of the Hessian to determine concavity.

##Stochastic Methods

Stochastic Optimization is used when…
*  the input or data is random or unpredictable (like how long it takes to travel somewhere)
*  the input is prone to measurement error (like scientific data from imperfect instruments)
*  the data or equations are known, but too complex to handle conventionally (like finding the probability of winning at solitaire)

All stochastic processes involve a large amount of input or trials; in general, the more trials there are, the more accurate the results will be.

The primary disadvantage of stochastic methods is that their accuracy is not very good, though it’s usually close enough. For this reason they are typically not used when another method is feasible.

The other disadvantage of stochastic methods is without computer assistance, they are slow. This is why they have only become widely used in the computer era.

*  **Monte Carlo Methods** are used when the inputs involve a large amount of uncertainty, or when several factors depend on each other. 

*  **Simulated Annealing** is used for functions with many local extremes. 

*  **Genetic Algorithms** attempt to find optimal solutions by using concepts related to the process of natural selection: inheritance, mutation, selection, crossover.
