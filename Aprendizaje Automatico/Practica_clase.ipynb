```python
import numpy as np

y = np.array([124,95,71,45,18])
Y = y.reshape(-1,1)
x = np.array([49,69,89,99,109])
x = x.reshape(-1,1)


#concatenacion bias
X = np.hstack((np.ones((x.shape[0],1)), x))


print(X.shape, Y.shape)
```

```python
#aplicacion del MSE
XT = X.T 

theta = np.dot(np.linalg.inv(np.dot(XT,X)),np.dot(XT,Y))
print(theta)
```

```python
import matplotlib.pyplot as plt

y_pred = np.dot(X,theta)
plt.plot(y_pred,'*')
plt.plot(Y,'.r')
```
